# Transform
Utiliza a bilioteca pandas para filtrar os dados necessários


```python
Country = df['Country'].tolist()
Country.sort()
print(Country[0:20])
```

```python
df.sort_values(by=['Country','Name'])
```

```python
count = df["Country"].value_counts()
count
```

```python
index_meu = df['Index'].count() +1
print(index_meu)
new_row =pd.DataFrame( {'Index': [index_meu],'Organization ID': ["8a9490A50adz241"], 'Name':["Luiz Fillipe"], 'Website' :["https://github.com/luizFillipe1"],'Country': ["Brazil"],'Description' :["Digital Innovation One"],'Founded':["2023"],'Industry' :["Other Industry"],'Number of employees':["500"]})
new_row

#Index,Organization Id,Name,Website,Country,Description,Founded,Industry,Number of employees/
df = pd.concat([new_row,df],join='inner',ignore_index=False)
df
```
